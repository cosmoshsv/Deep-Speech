```
import os
from datetime import datetime
from tqdm import tqdm
## DeepSpeech Simple

def flac2wav(folder):
    start_point = datetime.now()
    count_files = 0
    for dirpath, _, files in tqdm(os.walk(folder)):
#        print(dirpath)
        for a_file in files:
            if ".flac" in a_file:

                count_files += 1
#                print(dirpath, a_file, a_file.replace(".flac", ".wav"))
                os.system("sox " + dirpath + "/" + a_file + " " + dirpath + "/" + a_file.replace(".flac", ".wav"))
                print(a_file)
    print("folder: {}, data: {}\nprocessing time: {}".format(folder, count_files, datetime.now()-start_point))


folder = '/content/LibriSpeech/train-clean-100/'
flac2wav(folder=folder)
```

```
import glob, os
## Getting all the text files

wav_file = []

for root, dirs, files in os.walk("/content/LibriSpeech/train-clean-100"):
    for file in files:
        if file.endswith(".wav"):
             wav_file.append(os.path.join(root, file))
```

```
def process_file(file):
    file_toks = file.split('.')
    base_file_path = os.path.dirname(file)
    with open(file, 'r') as fd:
        lines = fd.readlines()
        for line in lines:
            toks = line.split(' ')
            wav_file_name = base_file_path + '/' + toks[0] + '.txt'
            print('Writing %s ' % wav_file_name)
            with open(wav_file_name, 'w') as of:
                trans = " ".join([t.lower() for t in toks[1:]])
                of.write(trans)
                of.write('\n')
```

```
libri_path = "/content/LibriSpeech/train-clean-100/"
```

```
transcript_files = [file for file in glob.glob(libri_path + "/*/*/*.txt")]
len(transcript_files)
```

```
for tf in transcript_files:
    process_file(tf)
```
